# Section 4: unsupervised learning 

The following analysis is centered on topics and how we can use those for clustering the business, using the similarity features vectors computes previously to know the topics on reviews, but in this case we group reviews by coffee shops. We don't want to classify the information depending on the connotation that the customer put on his review, instead how often people is talking about specific topics and how we can utilize that information to generate clusters of coffee.

Firstly, we used the similarity feature to convert rating column from string to float. Then we group the features by coffee shops with the aggregate function sum.We analyzed the features split into the original categories: coffee (beans, roast, baristas, drinks, sentiments), food (baked, breakfast, lunch, snacks, sentiments) and place (go, do, out, ambient, price, sentiments). To do that, hierarchical models were trained for building dendrograms and determine the number of clusters that allows splitting the coffee shops into reasonable groups and what criteria we can use to understand and interpret the results.

#### Importing the relevant packages:

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set(style="white")
```

Importing the Similarity Features Vectors:

```python
data1 = pd.read_csv('preprocessing_ml/similarity_features_vectors.csv')
```

```python
data1 = data1.set_index('coffee')
```

```python
data1['rating'] = data1['rating'].str.split(' star rating').str.get(0).astype(float)
```

```python
data1 = data1.groupby('coffee').sum()
```

```python
data1.loc[:, 'beans':'food_sentiment'].head(3)
```

#### Hierarchical Clustering using all the features:

In this section, we apply K-Means followed to Agglometive Clustering to find unsupervised groups of coffee shops. We implement a two-phase solution using k-Means with 12 clusters and then, as a second stage, a hierarchical clustering of a few less clusters (we'll try 3, 4 and 5).

```python
X = data1.loc[:, 'beans':'food_sentiment']
```

```python
clustering = KMeans(n_clusters=50, n_init=10,
random_state=1)
clustering.fit(X)
```

```python
Kx = clustering.cluster_centers_
Kx_mapping = {case:cluster for case,
 cluster in enumerate(clustering.labels_)}
```

```python
Hclustering = AgglomerativeClustering(n_clusters=3, affinity='cosine', linkage='complete')
Hclustering.fit(Kx)
```

```python
H_mapping = {case:cluster for case,
 cluster in enumerate(Hclustering.labels_)}
final_mapping = {case:H_mapping[Kx_mapping[case]] for case in Kx_mapping}
```

```python
ms = np.column_stack((data1.index,
 [final_mapping[n] for n in range(max(final_mapping)+1)]))
df = pd.DataFrame(ms, columns=['coffee','cluster'])
```

Projecting the clusters in a bidimensional space using dimensionality reduction PCA:

```python
pca = PCA(n_components=2)
points = pca.fit_transform(X)

df_pca = pd.DataFrame({'coffee': df.coffee,
                       'cluster_id': df.cluster.astype(float), 
                       'x': points[:, 0], 
                       'y': points[:, 1]})

df_pca.loc[df_pca.cluster_id == 0.0, 'category'] = 'A'
df_pca.loc[df_pca.cluster_id == 1.0, 'category'] = 'B'
df_pca.loc[df_pca.cluster_id == 2.0, 'category'] = 'C'

f, ax = plt.subplots(figsize=(6.5, 6.5))
_ = sns.despine(f, left=True, bottom=True)
_ = sns.scatterplot(x='x', y='y', hue='category', 
                    data=df_pca, palette='Set2')
_ = plt.xlabel('First principal component', fontsize=12)
_ = plt.ylabel('Second principal component', fontsize=12)
_ = plt.title('Coffee shops agglomerative clustering using all the features', fontsize=14)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

# Clustering Coffee Shops

Finally, can we distinguish clusters to separate the coffee shops. What kind of criteria we can use to split them?

The following analysis is centered on topics and how we can use those for clustering the business, using the similarity computes previously to know the topics on reviews, but in this case we group reviews by coffee shops. We don't want to classify the information depending on the connotation that the customer put on his review, but how often people is talking about specific topics and how we can utilize that information to generate clusters of coffee. 

### Pipeline

1. Use the file of similarity to convert rating column from string to float and replace the duplicate names of coffee shops.
2. Group by coffee shops with the aggregate function mean.
3. Analysis the features split into the original categories: **coffee** (beans, roast, baristas, drinks, sentiments), **food** (baked, breakfast, lunch, snacks, sentiments) and **place** (go, do, out, ambient, price, sentiments).


## Category: coffee

A dendogram of the coffee shops considering **only** sections related to coffee is displayed below.

```python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram

# All about coffee category
X = data1.loc[:, 'beans':'coffee_sentiment']

# Calculate the linkage: mergings
mergings = linkage(X, method='complete', metric='cosine')

f, ax = plt.subplots(figsize=(15, 10))
# Plot the dendrogram, using varieties as labels
dendrogram(mergings, labels=data1.index, leaf_rotation=90, leaf_font_size=7)
plt.show()
```

*Cosine* reduce the noise into accounting the shape of variables, more than their values and it is useful when you have many variables and you are not sure about the significance of them into the model. After to test variables using chi-square we know which variables have more influence for training models, but in this case we want to force that another variables take more protagonism and determine the label of the features vectors.

We extract labels inspecting the clusters generate at a distance of 0.25:

```python
from scipy.cluster.hierarchy import fcluster

# Use fcluster to extract labels: labels
labels = fcluster(mergings, 0.25, criterion='distance')

# Create a DataFrame with labels and varieties as columns: df
df_hier = pd.DataFrame({'labels': labels, 'varieties': data1.index})
df_hier.groupby('labels').count()
```

Looking at the dendogram we can find the cluster 3 that contains only one element to the left of the other branches. In practice, we are dealing with 5 clusters.

Reindexing data:

```python
data1_reindex = data1.reset_index()
```

And then applying PCA to get a bidimensional representation of the clusters:

```python
pca = PCA(n_components=2)
points = pca.fit_transform(X)

df_pca = pd.DataFrame({'coffee' : df_hier.varieties,
                       'cluster_id': df_hier.labels, 
                       'roast': data1_reindex.roast,
                       'x': points[:, 0], 
                       'y': points[:, 1]})

df_pca.loc[df_pca.cluster_id == 1, 'category'] = 'A'
df_pca.loc[df_pca.cluster_id == 2, 'category'] = 'B'
df_pca.loc[df_pca.cluster_id == 3, 'category'] = 'C'
df_pca.loc[df_pca.cluster_id == 4, 'category'] = 'D'
df_pca.loc[df_pca.cluster_id == 5, 'category'] = 'E'
df_pca.loc[df_pca.cluster_id == 6, 'category'] = 'F'

f, ax = plt.subplots(figsize=(6.5, 6.5))
_ = sns.despine(f, left=True, bottom=True)
_ = sns.scatterplot(x='x', y='y', hue='category',
                    palette='Set2', size='roast',
                    data=df_pca, hue_norm=(1, 6),
                    sizes=(20, 200))
_ = plt.xlabel('first principal component', fontsize=12)
_ = plt.ylabel('second principal component', fontsize=12)
_ = plt.title('Coffee clusters scaled by Roast feature', fontsize=14)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

The plot above is displaying coffee shops grouped in 5 clusters. The size of the elements depends on the sum of roasting expressions for every coffee business. We can observed, for example, that cluster F has the higher roasting sum, followed by the cluster B and D. The clusters with less alusion to the roasting of the seeds are E and A. Thus, information about **roasting** allows to distinguish categories of coffee shops and it is useful for the formation of clusters.

```python
df_pca[df_pca.category == 'E'].head()
```

```python
df_pca = pd.DataFrame({'coffee' : df_hier.varieties,
                       'cluster_id': df_hier.labels, 
                       'beans': data1_reindex.beans,
                       'x': points[:, 0], 
                       'y': points[:, 1]})

df_pca.loc[df_pca.cluster_id == 1, 'category'] = 'A'
df_pca.loc[df_pca.cluster_id == 2, 'category'] = 'B'
df_pca.loc[df_pca.cluster_id == 3, 'category'] = 'C'
df_pca.loc[df_pca.cluster_id == 4, 'category'] = 'D'
df_pca.loc[df_pca.cluster_id == 5, 'category'] = 'E'
df_pca.loc[df_pca.cluster_id == 6, 'category'] = 'F'

f, ax = plt.subplots(figsize=(6.5, 6.5))
_ = sns.despine(f, left=True, bottom=True)
_ = sns.scatterplot(x='x', y='y', hue='category', size='beans',
                    data=df_pca, palette='Set1',
                    sizes=(20, 200))
_ = plt.xlabel('first principal component', fontsize=12)
_ = plt.ylabel('second principal component', fontsize=12)
_ = plt.title('Coffee clusters scaled by Beans feature', fontsize=14)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
